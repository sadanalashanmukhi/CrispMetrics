# preparing dataset

```python
x=df.iloc[:,1:].values
y=df.iloc[:,0].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
len(x_test)
```

```python
acc=[]
model=[]
```

# training the models

# SVM classifier

```python
from sklearn.svm import SVC
import sklearn.metrics as metrics
SVM=SVC(gamma='auto')
SVM.fit(x_train,y_train)
predicted_values=SVM.predict(x_test)
x=metrics.accuracy_score(y_test,predicted_values)
acc.append(x)
model.append('SVM')
print("Test Accuracy : {:2f}%".format(SVM.score(x_test,y_test)*100))
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression()
LR.fit(x_train,y_train)
predictions=LR.predict(x_test)
x=LR.score(x_test,y_test)
acc.append(x)
model.append('LR')
print("Test Accuracy : {:2f}%".format(LR.score(x_test,y_test)*100))
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
RF=RandomForestClassifier()
RF.fit(x_train,y_train)
predictions=RF.predict(x_test)
r=LR.score(x_test,y_test)
acc.append(r)
model.append('RF')
print("Test Accuracy : {:2f}%".format(RF.score(x_test,y_test)*100))
```

# Decision tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
DT=DecisionTreeClassifier()
DT.fit(x_train,y_train)
predictions=DT.predict(x_test)
d=DT.score(x_test,y_test)
acc.append(d)
model.append('DT')
print("Test Accuracy : {:2f}%".format(DT.score(x_test,y_test)*100))
```

# Linear Discriminate Analysis

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
LDA=LinearDiscriminantAnalysis()
LDA.fit(x_train,y_train)
door=LDA.predict(x_test)
l=LDA.score(x_test,y_test)
acc.append(l)
model.append('LDA')
print("Test Accuracy : {:2f}%".format(LDA.score(x_test,y_test)*100))
```

# Accuracy of all algorithms

```python
a=print("SVM Classifier Accuracy : {:2f}%".format(SVM.score(x_test,y_test)*100))
b=print("Logistic Regression Accuracy : {:2f}%".format(LR.score(x_test,y_test)*100))
c=print("Random Forest Classifier Accuracy : {:2f}%".format(RF.score(x_test,y_test)*100))
d=print("Decision Tree Classifier Accuracy : {:2f}%".format(DT.score(x_test,y_test)*100))
e=print("Linear Discriminate Analysis Accuracy : {:2f}%".format(LDA.score(x_test,y_test)*100))
```

# Accuracy Comparison

```python
plt.figure(figsize=[10,5],dpi=100)
plt.title('Accuracy Comparison')
plt.xlabel('Algorithm')
plt.ylabel('Accuracy')
sns.barplot(x=model,y=acc,palette='dark')
```
